# ΑΣΚΗΣΗ 1 


```python

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import uniform , bernoulli


```

```python
def d(a,x):
    return a*x

def L(theta, d):
    return np.abs(theta - d)



```

```python
a=2**0.5 # παραδεκτη

a2=2 # αμεροληπτη
```

```python
N=1000
theta=10 
x=uniform(0,theta).rvs(N)

loss=0
loss2=0
plt.style.use('dark_background')
for i in range(N):
    loss+=L(theta,d(a,x[i]))
    loss2+=L(theta,d(a2,x[i]))
    plt.scatter(i,loss,c='b')
    plt.scatter(i,loss2,c='r')

    
    
    
    
    
    
```

# ΑΣΚΗΣΗ 2

```python
def d(a,x):
    return a*np.mean(x)

def L(theta, d):
    return (theta - d)**2

```

```python
N=100
theta=uniform(0,1).rvs(N)
```

```python
n=2
a=(2*n)/(2*n+1) # bayes μπλε
a2=1 # αμεροληπτη κοκκινη
loss=0
loss2=0
for i in range(N):
    x=bernoulli(theta[i]).rvs(n)
    loss+=L(theta[i],d(a,x))
    loss2+=L(theta[i],d(a2,x))
    plt.scatter(i,loss,c='b')
    plt.scatter(i,loss2,c='r')
    
```
